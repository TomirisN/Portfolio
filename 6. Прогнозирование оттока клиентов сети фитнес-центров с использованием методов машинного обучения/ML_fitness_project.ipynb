До сих пор мы рассматривали прогнозы модели как метки класса 1 или 0. На самом деле все немного сложнее. Статистические алгоритмы машинного обучения, такие как логистическая регрессия, основаны на вероятности; поэтому то, что на самом деле предсказывается двоичным классификатором, - это вероятность того, что метка истинна (P (y)) и вероятность того, что метка ложна (1 - P (y)). Пороговое значение 0,5 используется для определения того, является ли прогнозируемая метка равной 1 (P(y) > 0,5) или 0 (P(y) <= 0,5). Используем метод predict_proba, чтобы просмотреть пары вероятностей для каждого случая:

```python
y_scores = model.predict_proba(X_test)
print(y_scores)
```


Решение о присвоении прогнозу 1 или 0 баллов зависит от порогового значения, с которым сравниваются прогнозируемые вероятности. Если бы мы изменили пороговое значение, это повлияло бы на прогнозы и, следовательно, изменило бы показатели в confusion matrix. Распространенным способом оценки классификатора является проверка истинно положительного результата (что является другим названием отзыва) и ложноположительного результата для диапазона возможных пороговых значений. Затем эти показатели сопоставляются со всеми возможными пороговыми значениями, чтобы сформировать диаграмму, известную как received operator characteristic (ROC) chart, вот так:

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import confusion_matrix

# calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_scores[:,1])

# plot ROC curve
fig = plt.figure(figsize=(6, 6))
# Plot the diagonal 50% line
plt.plot([0, 1], [0, 1], 'k--')
# Plot the FPR and TPR achieved by our model
plt.plot(fpr, tpr)
plt.xlabel('Ложно положительная частота')
plt.ylabel('Истинно положительная частота')
plt.title('ROC кривая')
plt.show()
```

Диаграмма ROC показывает кривую частоты истинных и ложноположительных результатов для различных пороговых значений в диапазоне от 0 до 1. Идеальный классификатор имеет вид кривой, идущей прямо вверх по левой стороне и прямо поперек верхней (как в нашем случае). Диагональная линия поперек графика представляет вероятность правильного прогнозирования при случайном прогнозировании 50/50; таким образом, мы, очевидно, хотите, чтобы кривая была выше этой величины (или наша модель ничем не лучше простого угадывания!).

Площадь под кривой (AUC) - это значение в диапазоне от 0 до 1, которое количественно определяет общую производительность модели. Чем ближе к 1 это значение, тем лучше модель. Опять же, scikit-Learn включает функцию для вычисления этого показателя:

```python
from sklearn.metrics import roc_auc_score

auc = roc_auc_score(y_test,y_scores[:,1])
print('AUC: ' + str(auc))
```

AUC: 0.9779793993068925 - высокая производительность модели

### Выполнение предварительной обработки в паплайне - Логистическая регрессия

В этом случае кривая ROC и ее AUC указывают на то, что модель работает лучше, чем случайное предположение, что неплохо, учитывая, что мы выполнили очень небольшую предварительную обработку данных.

На практике обычно выполняется некоторая предварительная обработка данных, чтобы алгоритму было проще подогнать под них модель. Существует огромный спектр преобразований предварительной обработки, которые вы можете выполнить, чтобы подготовить свои данные к моделированию, но мы ограничимся несколькими распространенными методами:

- Масштабирование числовых объектов, чтобы они были в одном масштабе. Это предотвращает появление коэффициентов, непропорционально влияющих на прогнозы, при использовании функций с большими значениями.
- Кодирование категориальных переменных. Например, используя метод однократного горячего кодирования, вы можете создавать отдельные двоичные (true / false) функции для каждого возможного значения категории.

Чтобы применить эти преобразования предварительной обработки, мы воспользуемся функцией Scikit-Learn, называемой паплайнами. Они позволяют нам определить набор шагов предварительной обработки, которые заканчиваются алгоритмом. Затем вы можете подогнать весь паплайн к данным, чтобы модель инкапсулировала все этапы предварительной обработки, а также алгоритм регрессии. Это полезно, потому что, когда мы хотим использовать модель для прогнозирования значений на основе новых данных, нам нужно применить те же преобразования (на основе тех же статистических распределений и кодировок категорий, которые использовались с обучающими данными)

Паплайн инкапсулирует этапы предварительной обработки, а также обучение модели.

Далее используем модель, подготовленную этим паплайном, для прогнозирования меток для нашего набора тестов и сравним показатели производительности с базовой моделью, которую мы создали ранее.

```python
# Импорт необходимых библиотек
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score, 
                             roc_auc_score, confusion_matrix, roc_curve)
import time
```

### Предобработка данных

```python
# Определение предобработки для числовых и категориальных признаков
numeric_features = [7, 8, 9, 10, 11, 12]
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categorical_features = [0, 1, 2, 3, 4, 5, 6]
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)
```

### Определение функции для обучения и оценки моделей

```python
# Словарь для хранения результатов моделей
results = {}
best_params = {}
best_scores = {}
prediction_times = {}
y_scores_dict = {}

# Функция для обучения и оценки моделей
def train_and_evaluate(model_name, pipeline, param_grid):
    global results, prediction_times, best_params, best_scores, y_scores_dict

    # Подбор гиперпараметров
    grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid,
                               scoring='accuracy', cv=5, n_jobs=-1, verbose=1)

    model = grid_search.fit(X_train, y_train)

    # Сохранение результатов
    predictions = model.predict(X_test)
    y_scores = model.predict_proba(X_test)

    # Время прогнозирования
    start_time = time.time()
    _ = model.predict(X_test)
    end_time = time.time()
    prediction_times[model_name] = end_time - start_time

    # Сохранение результатов
    results[model_name] = [
        accuracy_score(y_test, predictions),
        precision_score(y_test, predictions, average='macro'),
        recall_score(y_test, predictions, average='macro'),
        roc_auc_score(y_test, y_scores[:, 1])
    ]

    # Сохранение лучших параметров и лучших кросс-валидируемых баллов
    best_params[model_name] = grid_search.best_params_
    best_scores[model_name] = grid_search.best_score_

    # Сохранение вероятностей для ROC-кривой
    y_scores_dict[model_name] = y_scores[:, 1]
```

### Логистическая регрессия

```python
# Логистическая регрессия
logistic_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('logregressor', LogisticRegression(solver='liblinear'))
])

logistic_param_grid = {
    'logregressor__C': [0.01, 0.1, 1, 10, 100],  
    'logregressor__penalty': ['l1', 'l2']
}

train_and_evaluate('Logistic Regression', logistic_pipeline, logistic_param_grid)

# Построение ROC-кривой для Логистической регрессии
fpr, tpr, _ = roc_curve(y_test, y_scores_dict['Logistic Regression'])
plt.figure(figsize=(10, 5))
plt.plot(fpr, tpr, label=f'Logistic Regression (AUC = {results["Logistic Regression"][3]:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.title('ROC-кривая для логистической регрессии')
plt.xlabel('Ложноположительная ставка')
plt.ylabel('Истинноположительная ставка')
plt.legend()
plt.grid()
plt.show()
```

#### Сравнительный анализ моделей

- **Качество предсказания**: Наилучшие показателиточности и AUC показаны у логистической регрессии. SVM также демонстрирует высокие результаты, но время прогнозирования у обеих моделей минимально.
- **Скорость прогнозирования**: Все модели выполняются быстро, однако логистическая регрессия оказывается наиболее быстрой, что делает её предпочтительным вариантом в условиях ограниченных ресурсов.
- **Интерпретируемость**: Логистическая регрессия имеет наивысшую интерпретируемость, поскольку коэффициенты могут напрямую быть связаны с вероятностями классов. Mетод опорных векторов и случайный лес менее интерпретируемы; однако, правая динамика для KNN среди всех кажется самой низкой из-за "чёрного ящика".
- **Визуализация**: Логистическая регрессия и KNN наиболее просто визуализируются из-за их простой структуры и глубины. Случайный лес является более сложным в этом аспекте.

#### Заключение

Выбор лучшей модели зависит от конкретных требований задачи. Если задача требует высокой интерпретируемости и скорости, **логистическая регрессия** будет наилучшим выбором. **Метод опорных векторов** предлагает гибкость и обнаружение сложных зависимостей, что делает его сильным кандидатом при наличии больших объемов данных. Случайный лес рекомендуется для задач, где стабильность и качество важнее интерпретируемости. **KNN** может обеспечить стабильные результаты на небольших наборах данных, но его масштабируемость ограничена.
